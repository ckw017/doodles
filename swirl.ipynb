```python
import drawSvg as draw
import random
from math import pi, sin, cos
from opensimplex import OpenSimplex
```

```python
gen = OpenSimplex(123)
```

```python
def woo(spacing=.3,noise_spin=1.1, noise_coeff=3, noise_scale = .02, dtheta=.05):
    canvas = draw.Drawing(1000, 1000, origin='center', displayInline=False)
    line = []
    theta = 0
    def radius(theta):
        return theta * spacing
    def noise(theta):
        theta *= noise_spin
        x = radius(theta) * cos(theta) * noise_scale
        y = radius(theta) * sin(theta) * noise_scale
        return gen.noise2d(x,y) * noise_coeff
    while radius(theta) < 400:
        r = radius(theta)
        r += noise(theta)
        line.append(r * cos(theta))
        line.append(r * sin(theta))
        theta += dtheta
    canvas.append(draw.Lines(*line, stroke='black', fill='none'))
    canvas.setPixelScale(.5)
    display(canvas.rasterize())
```

```python
woo()
```

```python
woo(noise_spin=1.01)
```

```python
woo(noise_spin=.99)
```

```python
woo(noise_spin=.99, spacing=.6)
```
