```python
import drawSvg as draw
from opensimplex import OpenSimplex
from math import pi, sin, cos

gen = OpenSimplex(seed=123)
```

```python
def squarez(depth, uwu=0, scale=0.5):
    canvas = draw.Drawing(800, 800, center=(0,0), displayInline=False)
    def square(x,y,d,rot,depth, exclude=[],scale=0.5):
        if depth == 0:
            return
        points = []
        corners = []
        for i in range(4):
            ang = rot + 2*pi*i/4
            px = x + cos(ang) * d
            py = y + sin(ang) * d
            points.extend([px, py])
            corners.append((px, py))
        canvas.append(draw.Lines(*points, close=True,fill='none',stroke='black'))
        if 0 not in exclude:
            square(*corners[0],d*scale,rot-uwu,depth-1,exclude=[2])
        if 1 not in exclude:
            square(*corners[1],d*scale,rot-uwu,depth-1,exclude=[3])
        if 2 not in exclude:
            square(*corners[2],d*scale,rot-uwu,depth-1,exclude=[0])
        if 3 not in exclude:
            square(*corners[3],d*scale,rot-uwu,depth-1,exclude=[1])
        
    
    square(400, 400, 200, pi/4, depth, scale=scale)
    display(canvas.rasterize())
```

```python
squarez(6, 0, scale=.5)
```

```python
squarez(7, 0.04, scale=.5)
```
