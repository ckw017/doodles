```python
import drawSvg as draw
from itertools import permutations
from math import pi, sin, cos
```

```python
def woops(tup):
    return tuple(map(lambda x: round(x, 1), tup))

def uhm(n=9, l=15200):
    dtheta = pi / n
    line_length = sin(dtheta/2)*500*2
    sorted_angles = [(n-1-i) * dtheta - (n-1)/2 * dtheta for i in range(n)]
    seen = set()
    uhm = 0
    d = draw.Drawing(1000, 1000, origin='center', displayInline=False)
    for perm in permutations(sorted_angles):
        if uhm % l != 0:
            uhm+=1
            continue
        uhm += 1
        start = (-500, 0)
        for starti in range(n):
            i = starti
            curr = start
            for _ in range(n):
                dx, dy = cos(perm[i]) * line_length, sin(perm[i]) * line_length
                end = (curr[0] + dx, curr[1] + dy)
                line = curr + end
                if woops(line) not in seen:
                    seen.add(woops(line))
                    d.append(draw.Line(*line, stroke='black'))
                curr = end
                i = (i + 1) % n
    print(len(seen))
    d.setPixelScale(0.5)
    display(d.rasterize())
```

```python
uhm()
```
