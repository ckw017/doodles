```python
import drawSvg as draw
from math import cos, sin, pi
```

```python
def lol(upper, rots, step_size=1):
    coeff = rots * (2 * pi) / upper
    d = draw.Drawing(650, 650, origin='center', displayInline=False)
    base_ang = 0
    r = upper
    while r >= 0:
        uwu = base_ang + coeff * r
        a = cos(uwu) * r, sin(uwu) * r
        b = cos(uwu + 2*pi/3) * r, sin(uwu + 2*pi/3) * r
        c = cos(uwu + 4*pi/3) * r, sin(uwu + 4*pi/3) * r
        d.append(draw.Lines(*a, *b, *c, close=True, stroke='black', fill='none'))
        r -= step_size
    d.setPixelScale(.5)
    display(d.rasterize())
```

```python
lol(300, 5)
```

```python
lol(300, 1)
```

```python
lol(200, 1/12, 4)
```
