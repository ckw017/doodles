```python
import drawSvg as draw
import random
from math import pi, sin, cos, asin
from opensimplex import OpenSimplex

gen = OpenSimplex(seed=123)
```

```python
def bark(start_r=0, end_r = 400, n=720, r_growth=2, noise_r_scale=0.02, noise_coeff_start=0, noise_coeff_growth=1):
    d = draw.Drawing(1000, 1000, origin='center', displayInline=False)
    dang = 2 * pi / n
    dr = r_growth / n
    ang = 0
    base_r, noise_r = start_r, start_r
    noise_coeff=noise_coeff_start
    dnc = noise_coeff_growth/n
    points = []
    while base_r < end_r:
        noise_x, noise_y = base_r * cos(ang), base_r * sin(ang)
        noisy_r = base_r + noise_coeff*gen.noise2d(noise_r_scale*noise_x, noise_r_scale*noise_y)
        x, y = noisy_r * cos(ang), noisy_r * sin(ang)
        points.extend((x, y))
        ang += dang
        base_r += dr
        noise_coeff += dnc
    print("oof?")
    d.append(draw.Lines(*points, close=False, fill='none', stroke='black'))
    print("oof!")
    d.setPixelScale(.5)
    display(d.rasterize())
# bark()
```

```python
def barc(start_r=320, end_r = 400, alen=60, r_growth=2, noise_r_scale=0.001, 
         noise_coeff_start=0, noise_coeff_growth=.1, hide=True, noisy_r_cap=None):
    d = draw.Drawing(1000, 1000, origin='center', displayInline=False)
    dr = r_growth
    ang = 0
    base_r, noise_r = start_r, start_r
    noise_coeff = noise_coeff_start
    points = []
    while base_r < end_r:
        noise_x, noise_y = base_r * cos(ang), base_r * sin(ang)
        hider = 0
        if hide:
            if base_r == start_r:
                hider = 2* r_growth
            if base_r + r_growth > end_r:
                hider = - 2 * r_growth
        noisy_r = base_r + hider + noise_coeff*gen.noise2d(noise_r_scale*noise_x, noise_r_scale*noise_y)
        if noisy_r_cap is not None and noisy_r > noisy_r_cap:
            noisy_r = noisy_r_cap
        x, y = noisy_r * cos(ang), noisy_r * sin(ang)
        points.extend((x, y))
        dang = 2 * asin(alen/2/base_r) 
        ang += dang
        base_r += dr * dang / (2*pi)
        noise_coeff += noise_coeff_growth * dang / (2*pi)
    print("oof?")
    d.append(draw.Lines(*points, close=False, fill='none', stroke='black'))
    print("oof!")
    d.setPixelScale(.5)
    display(d.rasterize())
barc()
```

```python

barc(start_r=220, end_r = 300, alen=2, r_growth=2, noise_r_scale=0.02, noise_coeff_start=2, noise_coeff_growth=.1, hide=False)
```
